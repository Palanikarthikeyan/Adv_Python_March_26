```python
class box:
    box_id = 101
    def f2(self):
        print(self.box_id) # same as obj.box_id  ---> 101
        self.box_id = 'B-101' # obj.box_id = 'B-101'

obj = box()
print(obj.box_id) # 101
obj.f2() # f2(obj)
print(obj.box_id) # B-101
```

```python
class box:
    box_id = 101
    def f2(self,a1):
        print(self.box_id) # same as obj.box_id  ---> 101
        self.box_id = a1 # obj.box_id = 'B-101'

obj = box()
print(obj.box_id) # 101
obj.f2('B-101') # f2(obj,B-101)
print(obj.box_id) # B-101
```

```python
## modify enrollment example
class Enrollment:
    emp_name = ''
    emp_dob = ''
    emp_dept = ''
    def f1 - initialization 
    def f2 - display emp records
    def f3 - update emp working dept - return updated dept details
```

```python
class Enrollment:
    emp_name = ''
    emp_dob = ''
    emp_dept = ''
    def f1(self,en,edob,edept):
        self.emp_name = en
        self.emp_dob = edob
        self.emp_dept = edept
        print(f'Emp {self.emp_name} enrollment is done!')
    def f2(self):
        print(f'About Emp {self.emp_name}  details:-')
        print(f'Emp name {self.emp_name} DOB:{self.emp_dob} Working dept is:{self.emp_dept}')
    def f3(self,edept):
        self.emp_dept = edept
        print(f'Emp {self.emp_name} working dept {self.emp_dept} is updated')
        print('updated details:-')
        self.f2() # nested methodCall
```

```python
eobj1 = Enrollment()
eobj1.f1('Arun','1st Jan','sales') # f1(eobj1,'Arun','1st Jan','sales')
```

```python
eobj2 = Enrollment()
eobj2.f1('Vijay','2nd Feb','Prod') # f1(eobj2,'Vijay','2nd Feb','Prod')
```

```python
eobj3 = Enrollment()
eobj3.f1('Anu','3rd March','HR') # f1(eobj3,'Anu','3rd March','HR')
```

```python
eobj1.f2()
```

```python
eobj2.f2()
```

```python
eobj3.f2()
```

```python
eobj3.f3('DBA')
```
