```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("altruistdelhite04/loan-prediction-problem-dataset")

print("Path to dataset files:", path)

train_df = pd.read_csv("/kaggle/input/datasets/altruistdelhite04/loan-prediction-problem-dataset/train_u6lujuX_CVtuZ9i.csv")
test_df = pd.read_csv("/kaggle/input/datasets/altruistdelhite04/loan-prediction-problem-dataset/test_Y3wMUE5_7gLdaTN.csv")
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
for a in train_df.columns:
    print(a)
    train_df[a] = train_df[a].fillna(train_df[a].mode()[0])
train_df.isnull().sum()
```

```python
for a in test_df.columns:
    print(a)
    test_df[a] = test_df[a].fillna(test_df[a].mode()[0])
test_df.isnull().sum()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
feature_cols = ['Gender', 'Married', 'Education','ApplicantIncome','LoanAmount','Credit_History']
X_train = train_df[feature_cols]
y_train = train_df['Loan_Status']
X_test= test_df[feature_cols]
```

```python
 from sklearn import preprocessing
 le = preprocessing.LabelEncoder()
 for column_name in train_df.columns:
        if train_df[column_name].dtype == object:
            train_df[column_name] = le.fit_transform(train_df[column_name])
        else:
            pass

print(train_df.describe)
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
for column_name in test_df.columns:
        if test_df[column_name].dtype == object:
            test_df[column_name] = le.fit_transform(test_df[column_name])
        else:
            pass
```

```python
logreg = LogisticRegression(random_state=42)
logreg.fit(X_train, y_train)
print("Training set:", X_train.shape)
print("Testing set:", X_test.shape)
```
